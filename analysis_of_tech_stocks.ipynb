## Import Packages

```python
import pandas as pd
import numpy as np
```

```python
import pandas_datareader as web
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Load the adjusted closings for the top 4 tech stocks

```python
# %pip install pandas_datareader yfinance
import yfinance as yf
```

```python
symbols = ["MSFT", "AMZN", "AAPL", "GOOG"]
start_date = "2019-01-01"
end_date = "2019-07-01"
stock_data_2019 = yf.download(symbols, start=start_date, end=end_date)
```

```python
stock_data_2019
```

```python
stock_data_2019['Adj Close']
```

```python
symbols = ["MSFT", "AMZN", "AAPL", "GOOG"]
start_date = "2022-01-01"
end_date = "2022-07-01"
stock_data_2022 = yf.download(symbols, start=start_date, end=end_date)
```

```python
stock_data_2022
```

```python
stock_data_2022['Adj Close']
```

## Calculate and visualize

```python
colors = ['blue', 'orange', 'green', 'red']
```

### 1. Plot the adjusted closing prices over time

```python
stock_data_closing_prices_2019 = stock_data_2019['Adj Close']
stock_data_closing_prices_2019.plot(color=colors)
plt.xlabel("Date")
plt.ylabel("Adjusted Closing Price")
plt.title("Tech Stocks Adjusted Price Over Time 2019")
plt.show()
```

```python
stock_data_closing_prices_2022 = stock_data_2022['Adj Close']
stock_data_closing_prices_2022.plot(color=colors)
plt.xlabel("Date")
plt.ylabel("Adjusted Closing Price")
plt.title("Tech Stocks Adjusted Price Over Time 2022")
plt.show()
```

Comparing the performance of the top four highest-valued technology stocks in the first half of 2019 to the first half of 2022, we observe significant growth and noteworthy actions taken by these companies during this period. These companies, which are leaders in the technology sector, have demonstrated impressive progress and strategic decision-making that contributed to their valuation growth.

### 2. Calculate and visualize the daily simple rate of return over time

```python
stock_data_daily_returns_2019 = stock_data_2019['Adj Close'].pct_change()
stock_data_daily_returns_2019.plot(color=colors)
plt.xlabel("Date")
plt.ylabel("ROR")
plt.title("Daily Simple Rate of Return Over time 2019")
plt.figure(figsize=(16,9))
plt.show()
```

```python
stock_data_daily_returns_2022 = stock_data_2022['Adj Close'].pct_change()
stock_data_daily_returns_2022.plot(color=colors)
plt.xlabel("Date")
plt.ylabel("ROR")
plt.title("Daily Simple Rate of Return Over time 2022")
plt.figure(figsize=(16,9))
plt.show()
```

In the first six months of 2019, Apple's daily returns exhibited higher volatility, reflecting increased instability relative to other companies. On the other hand, during the first half of 2022, Amazon experienced notable fluctuations in its daily returns compared to its industry counterparts, indicating a greater degree of oscillation. This suggests that both Apple and Amazon underwent periods of heightened market uncertainty and variability in their stock performance, potentially influenced by various factors such as economic conditions, industry dynamics, and company-specific events.

### 3. Create subplots of daily simple rate of return

```python
fig = plt.figure(figsize=(15,15))

ax1 = fig.add_subplot(321)
ax2 = fig.add_subplot(322)
ax3 = fig.add_subplot(323)
ax4 = fig.add_subplot(324)

ax1.plot(stock_data_2019['Adj Close']['AMZN'].pct_change(), color='orange')
ax1.set_title("Amazon")
ax2.plot(stock_data_2019['Adj Close']['AAPL'].pct_change(), color='b')
ax2.set_title("Apple")
ax3.plot(stock_data_2019['Adj Close']['GOOG'].pct_change(), color='g')
ax3.set_title("Google")
ax4.plot(stock_data_2019['Adj Close']['MSFT'].pct_change(), color='r')
ax4.set_title("Microsoft")

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(15,15))

ax1 = fig.add_subplot(321)
ax2 = fig.add_subplot(322)
ax3 = fig.add_subplot(323)
ax4 = fig.add_subplot(324)

ax1.plot(stock_data_2022['Adj Close']['AMZN'].pct_change(), color='orange')
ax1.set_title("Amazon")
ax2.plot(stock_data_2022['Adj Close']['AAPL'].pct_change(), color='b')
ax2.set_title("Apple")
ax3.plot(stock_data_2022['Adj Close']['GOOG'].pct_change(), color='g')
ax3.set_title("Google")
ax4.plot(stock_data_2022['Adj Close']['MSFT'].pct_change(), color='r')
ax4.set_title("Microsoft")

plt.tight_layout()
plt.show()
```

### 4. Calculate and plot the mean of each tech stock's daily simple rate of return

```python
daily_mean = stock_data_daily_returns_2019.mean()
daily_mean
```

```python
daily_mean.keys()
```

```python
height = []
for key in daily_mean.keys():
    height.append(daily_mean[key])
height
```

```python
x_pos = np.arange(len(daily_mean.keys()))
x_pos
```

```python
plt.bar(x_pos, height, color=colors, edgecolor='black')
 
plt.xticks(x_pos, daily_mean.keys())

plt.xlabel("Tech_Stocks")
plt.ylabel("daily mean")
plt.title("Daily mean rate of return 2019")

plt.show()
```

Microsoft has the highest mean simple rate of return over the period of data collected. Thus Microsoft would have been a good choice for investment over this period of time. Google, on the other hand, has the lowest mean simple rate of return over the period.

```python
daily_mean = stock_data_daily_returns_2022.mean()
daily_mean.keys()
height = []
for key in daily_mean.keys():
    height.append(daily_mean[key])
x_pos = np.arange(len(daily_mean.keys()))


plt.bar(x_pos, height, color=colors, edgecolor='black')
 
plt.xticks(x_pos, daily_mean.keys())

plt.xlabel("Tech_Stocks")
plt.ylabel("daily mean")
plt.title("Daily mean rate of return 2022")

plt.show()
```

All technology stocks had negative daily mean returns in the first half of 2022, it suggests that none of the stocks performed well during that period. This means that investing in any of the top four highest-valued technology stocks in the first half of 2022 may not have been a profitable choice.

### 7. Calculate the correlations
